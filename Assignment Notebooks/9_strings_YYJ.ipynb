### Stripping white space

Later we will learn how to process text files with Python. To do so, we can strip the white spaces so we can process the string. Those white spaces include
- ' ' or '\s'- spaces
- '\n' - new lines
- '\t' - tabs

```python
astring = "     Python is a versatile, readable language  .\t... \n"
# .strip() takes out all white spaces from the ends
print("original")
print(astring)
print("*")
print("stripped")
print(astring.strip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.lstrip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.rstrip())
print("*")
```

```python
def main():
    user_input = input("Enter a string to check if it's a palindrome: ")
    lowercase_input = user_input.lower()
    print(''.join(reversed(lowercase_input)))
    # Alternative to [::-1] - using reversed() and join()
    if lowercase_input == ''.join(reversed(lowercase_input)):
        print(f"'{user_input}' is a palindrome!")
    else:
        print(f"'{user_input}' is not a palindrome.")

    
main()

```

```python
def main():
    user_input = input("Enter a string to check if it's a palindrome: ")
    lowercase_input = user_input.lower()
    left, right = 0, len(lowercase_input) - 1

    while left < right:
        if lowercase_input[left] != lowercase_input[right]:
            return False
        left += 1
        right -= 1
    return True



    
main()

```

```python
class Dog:
    def __init__(self, name, color, breed, age):
        '''Initialize a Dog with name, color, breed, age'''
        self.name = name
        self.color = color
        self.breed = breed
        self.age = age
    
    def display_info(self):
        ''' Method that displays the information of the Dog '''
        print("Dog Information:")
        print("  Name:", self.name)
        print("  Color:", self.color)
        print("  Breed:", self.breed)
        print("  Age:", self.age, "years old")

    def change_age(self, age):
        ''' Method that allows the user to change the age of the dog '''
        self.age = age
        print(self.name, "'s age has been updated to", self.age, "years old")

def main():
    '''
    Ask the user to enter a dog name, color, breed, and age
    Create an instance of two dogs
    Use the display_info method to display the information
    Use the change_age method to change the age of one instance
    '''
    # Get information for first dog
    print("Enter information for Dog 1:")
    name1 = input("Enter dog name: ")
    color1 = input("Enter dog color: ")
    breed1 = input("Enter dog breed: ")
    age1 = int(input("Enter dog age: "))
    
    # Get information for second dog
    print("\nEnter information for Dog 2:")
    name2 = input("Enter dog name: ")
    color2 = input("Enter dog color: ")
    breed2 = input("Enter dog breed: ")
    age2 = int(input("Enter dog age: "))
    
    # Create dog instances
    dog1 = Dog(name1, color1, breed1, age1)
    dog2 = Dog(name2, color2, breed2, age2)
    
    # Display original information
    print("\nOriginal Dog Information")
    dog1.display_info()
    dog2.display_info()
    
    # Change age of one dog
    print("\nUpdating Age")
    new_age = int(input("Enter new age for dog1"))
    dog1.change_age(new_age)
    
    # Display updated information
    print("\nUpdated Dog Information")
    dog1.display_info()
    dog2.display_info()

# Run the program
if __name__ == "__main__":
    main()
```

```python
class Dog:
    def __init__(self, name, color, breed, age):
        '''Initialize a Dog with name, color, breed, age'''
        self.name = name
        self.color = color
        self.breed = breed
        self.age = age
    
    def display_info(self):
        ''' Method that displays the information of the Dog '''
        print(f"Dog Information:")
        print(f"  Name: {self.name}")
        print(f"  Color: {self.color}")
        print(f"  Breed: {self.breed}")
        print(f"  Age: {self.age} years old")
        print()  # Empty line for spacing

    def change_age(self, age):
        ''' Method that allows the user to change the age of the dog '''
        self.age = age
        print(f"{self.name}'s age has been updated to {self.age} years old")

def main():
    '''
    Ask the user to enter a dog name, color, breed, and age
    Create an instance of two dogs
    Use the display_info method to display the information
    Use the change_age method to change the age of one instance
    '''
    # Get information for first dog
    print("Enter information for Dog 1:")
    name1 = input("Enter dog name: ")
    color1 = input("Enter dog color: ")
    breed1 = input("Enter dog breed: ")
    age1 = int(input("Enter dog age: "))
    
    # Get information for second dog
    print("\nEnter information for Dog 2:")
    name2 = input("Enter dog name: ")
    color2 = input("Enter dog color: ")
    breed2 = input("Enter dog breed: ")
    age2 = int(input("Enter dog age: "))
    
    # Create dog instances
    dog1 = Dog(name1, color1, breed1, age1)
    dog2 = Dog(name2, color2, breed2, age2)
    
    # Display original information
    print("\n--- Original Dog Information ---")
    dog1.display_info()
    dog2.display_info()
    
    # Change age of one dog
    print("--- Updating Age ---")
    new_age = int(input(f"Enter new age for {dog1.name}: "))
    dog1.change_age(new_age)
    
    # Display updated information
    print("\n--- Updated Dog Information ---")
    dog1.display_info()
    dog2.display_info()

# Run the program
if __name__ == "__main__":
    main()
```
